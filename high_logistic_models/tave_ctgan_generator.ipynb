```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置专业绘图风格
sns.set_style("ticks")
plt.style.use("seaborn-v0_8-paper")

# 配色方案
main_color = "#2E5077"      # 深蓝色
reference_color = "#C96767" # 暗红色
marker_color = "#37718E"    # 蓝灰色

# 数据处理部分
csv_files = [f for f in os.listdir('.') if f.endswith('.csv') and 'n2-200_' in f]
x_values = []
y_values = []

for file in csv_files:
    x = int(file.split('n2-200_')[1].split('.csv')[0])
    if x <= 30:
        x = x/100 * 414
    else:
        x = ((x-30)*2+30)/100 * 414
        
    df = pd.read_csv(file)
    y = df.iloc[0, 2] if df.shape[0] == 1 else df.iloc[2, 6]
    
    x_values.append(x)
    y_values.append(y)

# 数据排序
sorted_pairs = sorted(zip(x_values, y_values))
x_values = [x for x, _ in sorted_pairs]
y_values = [y for _, y in sorted_pairs]

# 图表初始化（标准尺寸）
plt.figure(figsize=(14, 10))

# 设置纯白色背景
plt.gca().set_facecolor('#FFFFFF')
plt.gcf().set_facecolor('#FFFFFF')

# 主数据线（标准线宽）
plt.plot(x_values, y_values, 
         color=main_color, 
         linewidth=2.5,  # 恢复标准线宽
         linestyle='-',
         zorder=3)

# 数据点样式（标准尺寸）
plt.scatter(x_values, y_values,
           color=marker_color,
           s=50,  # 标准数据点大小
           zorder=4,
           label='Generated Data',
           edgecolor='white',
           linewidth=2)

# 参考线设置（标准线宽）
plt.axhline(y=0.24074, 
           color=reference_color, 
           linestyle='--', 
           linewidth=2.0,  # 恢复标准线宽
           label='Original Data',
           alpha=0.8,
           zorder=2)

# 标题和标签设置（标准字体）
plt.title('Prediction Error of Generated Data(Logistic,$C_0=2$)', 
         fontsize=26,        # 标准标题字号
         pad=30,             # 标准标题间距
         fontfamily='serif') # 移除加粗参数

plt.xlabel('Number of Source Data Sets', 
         fontsize=22,        # 标准标签字号
         labelpad=15,        # 标准标签间距
         fontfamily='serif')

plt.ylabel('Prediction Error', 
         fontsize=22,
         labelpad=15,
         fontfamily='serif')

# 网格设置
plt.grid(True, linestyle='--', alpha=0.3, zorder=1)

# 刻度参数（标准设置）
sns.despine(offset=10, trim=False)
plt.tick_params(
    axis='both',
    labelsize=20,            # 标准刻度字号
    length=6,                # 标准刻度长度
    width=1.5,               # 标准刻度宽度
    pad=10                   # 标准标签间距
)

# 显示范围设置
plt.ylim(0.225, 0.26)  # 保持手动设置

# 图例设置（标准字体）
legend = plt.legend(loc='best',
                   frameon=True, 
                   fontsize=18,      # 标准图例字号
                   fancybox=True,
                   framealpha=0.95,
                   edgecolor='none')

# 图例背景设置
legend.get_frame().set_facecolor('#FFFFFF')

# 保存设置
plt.tight_layout()
plt.savefig(r'C:\Users\jiang\Desktop\展示图片\sigmoid.pdf', 
           dpi=600,          # 标准分辨率
           bbox_inches='tight',
           facecolor='#FFFFFF',
           transparent=False)

plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置专业绘图风格（来自模板）
sns.set_style("ticks", {"axes.linewidth": 1.2, "axes.edgecolor": "#333333"})
plt.style.use("seaborn-v0_8-paper")
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20
plt.rcParams['axes.labelpad'] = 15
plt.rcParams['xtick.major.pad'] = 10
plt.rcParams['ytick.major.pad'] = 10

# 配色方案（来自模板）
main_color = "#1F4E79"      # 深蓝色，Ours
ctgan_color = "#A83333"     # 暗红色，CTGAN
tvae_color = "#4A7043"      # 深绿色，TVAE
ours_unfiltered_color = "#2A6A4D"  # 深绿色，Ours(unfiltered)
reference_color = "#D98C8C" # 参考线颜色

# 数据处理函数（增加两段式逻辑选项和 x <= 300 过滤）
def load_data(directory, prefix, row=0, col=0, two_stage=False):
    files = sorted([f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith('.csv')])
    xs, ys = [], []
    
    if two_stage:  # CTGAN 和 TVAE 使用两段式逻辑
        seq1 = [i / 100 * 414 for i in range(1, 31)]  # 1%-30%, 步长 0.01
        seq2 = [(0.4 + (i - 31) * 0.1) * 414 for i in range(31, 38)]  # 40%-100%, 步长 0.1 (31-37)
        x_values = seq1 + seq2
    else:  # Ours 和 Ours(unfiltered) 使用三段式逻辑
        seq1 = [i / 100 * 414 for i in range(1, 31)]  # 1%-30%, 步长 0.01
        seq2 = [((i - 30) * 0.02 + 0.30) * 414 for i in range(31, 41)]  # 32%-50%, 步长 0.02
        seq3 = [((i - 40) * 0.1 + 0.5) * 414 for i in range(41, 46)]  # 60%-100%, 步长 0.1
        x_values = seq1 + seq2 + seq3

    for file in files:
        base = os.path.basename(file)
        try:
            num = int(base.split('_')[-1].split('.')[0])
            if two_stage and (num < 1 or num > 37):  # 两段式范围 1-37
                print(f"Skipping {file}: index {num} out of range for two-stage")
                continue
            elif not two_stage and (num < 1 or num > 45):  # 三段式范围 1-45
                print(f"Skipping {file}: index {num} out of range for three-stage")
                continue
            x_val = x_values[num - 1]  # 从 x_values 中取值
            if x_val > 300:  # 过滤掉 x > 300 的点
                print(f"Skipping {file}: x={x_val} > 300")
                continue
        except Exception as e:
            print(f"Error parsing {file}: {e}")
            continue
        
        df = pd.read_csv(os.path.join(directory, file))
        y_val = df.iloc[row, col]
        xs.append(x_val)
        ys.append(y_val)

    # 按 x 值排序，与 German 一致
    sorted_pairs = sorted(zip(xs, ys))
    xs = [x for x, y in sorted_pairs if y is not None]
    ys = [y for _, y in sorted_pairs if y is not None]
    print(f"{prefix} data: x={xs}, y={ys}")
    return xs, ys

# 读取数据
# 1. Ours
ours_dir = '.'
ours_x, ours_y = load_data(ours_dir, "n2-200_", row=0, col=2 if pd.read_csv(os.path.join(ours_dir, "n2-200_1.csv")).shape[0] == 1 else 6, two_stage=False)

# 2. Ours(unfiltered)
ours_unfiltered_dir = r'C:\Users\jiang\Desktop\备份\high_logistic\all'
ours_unfiltered_x, ours_unfiltered_y = load_data(ours_unfiltered_dir, "highlogall_", row=0, col=0, two_stage=False)

# 3. CTGAN
ctgan_dir = r'C:\Users\jiang\Desktop\备份\high_logistic\ctgan'
ctgan_x, ctgan_y = load_data(ctgan_dir, "ctgan_", row=0, col=0, two_stage=True)

# 4. TVAE
tvae_dir = r'C:\Users\jiang\Desktop\备份\high_logistic\tvae'
tvae_x, tvae_y = load_data(tvae_dir, "tvae_", row=0, col=0, two_stage=True)

# 作图
fig, ax = plt.subplots(figsize=(12, 8), facecolor='white')
ax.set_facecolor('white')

# 数据线与标记（使用模板样式）
# Ours: 实线 + 五角星
ax.plot(ours_x, ours_y, color=main_color, linewidth=2, linestyle='-', marker='*', markersize=8, 
        markeredgecolor=main_color, markeredgewidth=1, zorder=3, alpha=0.9, label='Ours')

# Ours(unfiltered): 实线 + 圆形
ax.plot(ours_unfiltered_x, ours_unfiltered_y, color=ours_unfiltered_color, linewidth=2, linestyle='-', 
        marker='o', markersize=6, markeredgecolor='white', markeredgewidth=0.5, zorder=3, alpha=0.9, 
        label='Ours(unfiltered)')

# CTGAN: 虚线 + 方形
ax.plot(ctgan_x, ctgan_y, color=ctgan_color, linewidth=2, linestyle='--', marker='s', markersize=6, 
        markeredgecolor='white', markeredgewidth=0.5, zorder=3, alpha=0.9, label='CTGAN')

# TVAE: 双点划线 + 叉
ax.plot(tvae_x, tvae_y, color=tvae_color, linewidth=2, linestyle='--', marker='x', markersize=6, 
        markeredgecolor=tvae_color, markeredgewidth=2, zorder=3, alpha=0.9, label='TVAE')

# 参考线
ax.plot([0, 290], [0.24074, 0.24074], color=reference_color, linestyle='--', linewidth=1.5, label='Original Data', 
        alpha=0.7, zorder=2)

# 标题和标签
ax.set_title('Prediction Error of Generated Data (Logistic)', fontsize=24, pad=20, weight='bold')
ax.set_xlabel('Number of Source Data Sets', fontsize=20)
ax.set_ylabel('Prediction Error', fontsize=20)

# 优化坐标轴
sns.despine(ax=ax, offset=10, trim=False)
ax.tick_params(axis='both', labelsize=18, length=6, width=1, color='#333333')
ax.set_xlim(left=0, right=300)  # 设置 x 轴范围上限为 300

# 自动调整 y 轴范围并添加边距
current_ylim = ax.get_ylim()
y_margin = (current_ylim[1] - current_ylim[0]) * 0.1
ax.set_ylim(current_ylim[0] - y_margin, current_ylim[1] + y_margin)

# 添加网格
ax.grid(True, linestyle='--', linewidth=0.5, color='#D3D3D3', alpha=0.6, zorder=1)

# 图例
legend = ax.legend(loc='upper right', fontsize=14, frameon=True, fancybox=True, framealpha=0.9, edgecolor='none')
legend.get_frame().set_facecolor('white')
legend.get_frame().set_linewidth(0)


caption = r"""Figure 6:Prediction Error Comparison on High-Dimensional Logistic Regression Simulation. "Ours" refers to data generated using Glmtrans with filtering.This figure highlights the necessity of filtering."""

plt.figtext(0.5,                  # x position (center)
            -0.03,                 # y position (near bottom) 
            caption,
            ha='center',
            va='bottom',
            fontsize=14,
            color='#333333',
            wrap=True,
            bbox=dict(facecolor='white', 
                     edgecolor='none', 
                     pad=5))

# 调整布局并保存
plt.tight_layout()
plt.savefig(r'C:\Users\jiang\Desktop\rebuttal\plot\Fig_6_High_Logistic.pdf', dpi=600, bbox_inches='tight', facecolor='white', 
            transparent=True, format='pdf')
plt.show()
```
